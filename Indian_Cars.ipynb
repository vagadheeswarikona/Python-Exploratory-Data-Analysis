#Pie_Chart

```python
def small_percentages(data_series, threshold_percent=10):
    percentages = data_series.value_counts(normalize=True) * 100
    mask = percentages >= threshold_percent
    grouped_series = data_series.apply(lambda x: x if mask.get(x, False) else 'Others')
    return grouped_series
```

The small_percentages function groups values in a data_series that occur less than threshold_percent into 'Others'. It calculates the percentage of each unique value and replaces those below the threshold with 'Others'.

```python
# Remove 'unspecified' values from each column
car_specified = data.replace('unspecified', np.nan).dropna()
# Create a 2x2 matrix for pie charts
plt.figure(figsize=(10, 8))
# Pie chart for Drivetrain
plt.subplot(2, 2, 1)
drivetrain_distribution = car_specified['drivetrain'].value_counts()
plt.pie(drivetrain_distribution, labels=drivetrain_distribution.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Drivetrain Types')
plt.legend(loc='lower left', fontsize='small')

# Pie chart for Emission Norm
plt.subplot(2, 2, 2)
emission_norm_distribution = car_specified['emission_norm'].value_counts()
plt.pie(emission_norm_distribution, labels=emission_norm_distribution.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Emission Norms')
plt.legend(loc='lower left', fontsize='small')

# Pie chart for Fuel Type
plt.subplot(2, 2, 3)
fuel_type_distribution = small_percentages(car_specified['fuel_type'])
plt.pie(fuel_type_distribution.value_counts(), labels=fuel_type_distribution.value_counts().index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Fuel Types')
plt.legend(loc='lower left', fontsize='small')

# Pie chart for Desired Body Types
plt.subplot(2, 2, 4)
body_type_distribution = small_percentages(filtered_car['body_type'])
plt.pie(body_type_distribution.value_counts(), labels=body_type_distribution.value_counts().index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Desired Body Types in the Market')
plt.legend(loc='lower left', fontsize='small')

plt.tight_layout()
plt.show()
```
